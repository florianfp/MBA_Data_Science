```python
from sklearn import tree
from sklearn import ensemble
from sklearn.externals.six import StringIO 
import pydot 
import pandas as pd
import numpy as np
import csv
```

```python
#I/O
train_file = 'train.csv'
df = pd.read_csv(train_file)

test_file = 'test.csv'
df_test = pd.read_csv(test_file)
```

```python
def data_wrangling_train(data_frame):
    df = data_frame
    
    #transforming all features into numeric
    df['Sex2'] = df['Sex'].map({'female':0,'male':1})
    df['Embarked2'] = df['Embarked'].map({'C':0, 'Q':1, 'S':2})
    
    #DECISION 1 : How to fill in the NAs in the Age column
    #SOLUTION 1 : Take the median of the associated class for the given sex of the person
    #Let's remark that this solution is sub-optimal since we are only using the training set for this average and not the 
    #testing set
    #find those median ages
    median_ages = np.zeros((2,3)) #lines are for a same sex, columns are for a same class
    for i in range(0,2):
        for j in range(0,3):
            median_ages[i,j] = df[(df['Sex2']==i) & (df['Pclass']==j+1)]['Age'].dropna().median()
            
    #before filling in Age values, keep track of missing Ages
    df['AgeMissing'] = df['Age'].isnull().astype(int)

    #replace NAs Ages by median ages
    for i in range(0,2):
        for j in range(0,3):
            df.loc[(df['Age'].isnull()) & (df['Sex2']==i) & (df['Pclass']==j+1),'Age'] = median_ages[i,j]
    
    #create the family size additional feature
    df['FamilySize'] = df['SibSp'] + df['Parch']
    
    #drop columns that are not useful
    df = df.drop(['Name', 'Sex', 'Ticket', 'Cabin', 'Embarked','AgeMissing'],axis=1)
    #drop NAs in remaining features
    df = df.dropna()
    
    return(df)
```

```python
def data_wrangling_test(data_frame):
    """A slight variation from the data wrangling for test given that no rows can be dropped since we want prediction for 
    each row of the imput
    """
    df = data_frame
    
    #transforming all features into numeric
    df['Sex2'] = df['Sex'].map({'female':0,'male':1})
    df['Embarked2'] = df['Embarked'].map({'C':0, 'Q':1, 'S':2})
    
    #DECISION 1 : How to fill in the NAs in the Age column
    #SOLUTION 1 : Take the median of the associated class for the given sex of the person
    #Let's remark that this solution is sub-optimal since we are only using the training set for this average and not the 
    #testing set
    #find those median ages
    median_ages = np.zeros((2,3)) #lines are for a same sex, columns are for a same class
    for i in range(0,2):
        for j in range(0,3):
            median_ages[i,j] = df[(df['Sex2']==i) & (df['Pclass']==j+1)]['Age'].dropna().median()
            
    #before filling in Age values, keep track of missing Ages
    df['AgeMissing'] = df['Age'].isnull().astype(int)

    #replace NAs Ages by median ages
    for i in range(0,2):
        for j in range(0,3):
            df.loc[(df['Age'].isnull()) & (df['Sex2']==i) & (df['Pclass']==j+1),'Age'] = median_ages[i,j]
            
    #DECISION 2 : How to fill in the NAs in the Fare column
    #SOLUTION 2 : Take the median of the associated class for the given class
    #Let's remark that this solution is sub-optimal since we are only using the training set for this average and not the 
    #testing set
    #find those median ages
    median_fares = np.zeros((1,3)) #lines are for a same sex, columns are for a same class
    for i in range(0,3):
        median_fares[0,i] = df[(df['Pclass']==i+1)]['Fare'].dropna().median()
    median_fares
    
    #before filling in Age values, keep track of missing Ages
    df['FareMissing'] = df['Fare'].isnull().astype(int)

    #replace NAs Ages by median ages
    for i in range(0,3):
        df.loc[(df['Fare'].isnull()) & (df['Pclass']==i+1),'Fare'] = median_ages[0,i]
    
    #create the family size additional feature
    df['FamilySize'] = df['SibSp'] + df['Parch']
    
    #drop columns that are not useful
    df = df.drop(['Name', 'Sex', 'Ticket', 'Cabin', 'Embarked','AgeMissing','FareMissing'],axis=1)
    #drop NAs in remaining features
    df = df.dropna()
    
    return(df)
```

```python
#training data
train_df = data_wrangling_train(df)
train = train_df.values

#testing data
test_df = data_wrangling_test(df_test)
test= test_df.values
```

```python
#Preparing the data for the algorithms
X = train[:,2:len(train)]
Y = train[:,1]
test = test[:,1:len(test)]
ids = test_df['PassengerId']
test_df.head(5)
```

```python
#Simple Classification Tree
clf = tree.DecisionTreeClassifier()
clf = clf.fit(X, Y)
```

```python
#getting the importance of every feature in the classification tree
feature_importance = clf.feature_importances_
columns_name = train_df.columns[2:]

df_features = pd.DataFrame(data=columns_name, columns=['Feature'])
df_features['FeatureImportance'] = feature_importance
df_features
```

```python
output = clf.predict(test)
len(output)
```

```python
#writing the result to a csv file
df_result = pd.DataFrame(data=ids)
df_result['output'] = output
df_result.info()

predictions_file = open("classifierTree.csv", "wb")
open_file_object = csv.writer(predictions_file)
open_file_object.writerow(["PassengerId","Survived"])
open_file_object.writerows(zip(ids, output))
predictions_file.close()
```
